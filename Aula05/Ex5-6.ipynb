```python
import skimage
import numpy as np
from matplotlib import pyplot as plt

image = skimage.io.imread("Pavlovsk_Railing_of_bridge_Yellow_palace_Winter.jpg")
```

Detectando se a imagem está sub/super exposta, analisado o histograma.

Verifica o histograma, procurando identificar se as regiões com maior intensidade (brilho), estão cortadas (superexposta), ou com frequências muito baixas (subexposta)

De maneira similar pode-se utilizar a função rescale_intensity, do modulo exposure do scikit-image para cortar parte do histograma, ou alargar tal

```python
grayscale_image = skimage.color.rgb2gray(image)

img_underexposed = skimage.exposure.rescale_intensity(grayscale_image, out_range=(0,100))

histogram, histogram_centers = skimage.exposure.histogram(img_underexposed)

fig, ax = plt.subplots(1, 2, figsize=(20, 5))
ax[0].imshow(img_underexposed, cmap=plt.cm.gray)
ax[0].axis("off")
ax[0].set_title("Image (grayscale)")
ax[1].plot(histogram_centers, histogram)
ax[1].set_title('Histogram')
```

Detectando se a imagem está com alto/baixo contraste, baseado no histograma

Verifica-se o histograma, para casos de que as intensidades com maior frequência estejam agrupados (formando um pico ), ou alargados, sendo o primeiro caso, normalmente referente a imagens com baixo contraste, e o segundo caso para imagens com alto contraste. 

Similarmente, pode-se usar o modulo exposure do scikit-image, como outrora feito, para que se utilize a função is_low_contrast, retornando um boolean para o resultado

```python
skimage.exposure.is_low_contrast(image)
```
